```python
import numpy as np
import matplotlib.pyplot as plt
from singledimga import SingleDimGA
# %matplotlib
```

```python
f = lambda x: x * np.sin(10 * np.pi * x) + 1
# f(x1,x2) = sin(10x1) +x1cos(2πx2), x1∈[−2,2],x2∈[0,1].
#f = lambda x: x

a = -2
b = 2
h = 1e-6
n = 100
```

```python
x = np.arange(a, b, h)
plt.plot(x, f(x))
```

```python
single_dim_ga = SingleDimGA()
res = single_dim_ga.solve(f, a=a, b=b, h=h, n=n,
                          tournament_n=10,
                          mutation_p=0.1,
                          crossbreeding_p=0.9,
                          max_iter=100,
                          max_no_conv_iter=20)

print(res)
```

```python
plt.plot(single_dim_ga.history)
plt.show()
```
