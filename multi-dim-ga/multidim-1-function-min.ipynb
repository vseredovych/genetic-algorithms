```python
import numpy as np
import matplotlib.pyplot as plt
from multidimga import MultiDimGA
# %matplotlib
```

```python
f = lambda x, y: np.sin(10*x) + x * np.cos(2*np.pi*y)

intervals = np.array([
    [0, 2], 
    [0, 2]
])
```

```python
x = np.linspace(intervals[0][0], intervals[0][1], 100)
y = np.linspace(intervals[1][0], intervals[1][1], 100)

xx, yy = np.meshgrid(x, y)
zz = f(xx, yy)
```

```python
fig = plt.figure(figsize=(10,8))
fig.add_subplot(1, 1, 1)
plt.contourf(xx, yy, zz)
plt.colorbar()
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1, 1, 1, projection = '3d')
ax.contour3D(xx, yy, zz, 180, cmap='viridis')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

```python
h = 1e-12
n = 50
tournament_n=5
mutation_p=0.2
crossbreeding_p=0.95
max_iter = 200
max_no_conv_iter = 20


multi_dim_ga = MultiDimGA()
x, y = multi_dim_ga.solve(f,
                         intervals=intervals, 
                         h=h, 
                         n=n,
                         tournament_n=tournament_n,
                         mutation_p=mutation_p,
                         crossbreeding_p=crossbreeding_p,
                         max_iter=max_iter,
                         max_no_conv_iter=max_no_conv_iter,
                         max=False)
```

```python
print(f"x1: {x[0]}\nx2: {x[1]}\ny:  {y}")
```

```python
plt.plot(multi_dim_ga.history)
plt.show()
```
